```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt

# Load the pre-trained Stable Diffusion model from Hugging Face
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4-original", torch_dtype=torch.float16)
pipe.to("cuda")  # Move the model to GPU for faster inference

# Function to generate image from a text prompt
def generate_image_from_prompt(prompt):
    # Generate an image using the Stable Diffusion model
    image = pipe(prompt).images[0]

    # Show the generated image
    plt.imshow(image)
    plt.axis('off')
    plt.show()

    return image

# Input: Text prompt (can modify to use as needed)
prompt = "A beautiful landscape with mountains and a river"

# Generate the image
generated_image = generate_image_from_prompt(prompt)
```

```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# Load the Stable Diffusion 2.1 model from Hugging Face
pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1", torch_dtype=torch.float16)

# Move the model to GPU for faster inference
pipe.to("cuda")

# Function to generate an image based on a text prompt
def generate_image_from_prompt(prompt):
    # Generate an image using the Stable Diffusion model
    image = pipe(prompt).images[0]

    # Show the generated image
    plt.imshow(image)
    plt.axis('off')
    plt.show()

    return image

# Example: Input prompt for generating an image
prompt = "cinema hall with crowd"

# Generate the image
generated_image = generate_image_from_prompt(prompt)
```
